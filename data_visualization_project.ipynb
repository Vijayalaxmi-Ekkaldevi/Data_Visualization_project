```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
data=pd.read_csv("netflix_titles.csv")
df=pd.DataFrame(data)
df
```

```python
df["director"]=df["director"].fillna("Not Available")
```

```python
df["cast"]=df["cast"].fillna("Not Available")
```

```python
df["country"]=df["country"].fillna("Unknown")
```

```python
df.dropna(subset=['date_added'], inplace=True)
```

```python
df["rating"]=df["rating"].fillna("Not Rated")
```

```python
df["duration"]=df["duration"].fillna("Unknown")
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
plt.figure(figsize=(8,4))
sns.countplot(x="type",data=df,palette=["#4E79A7", "#F28E2B"],hue="type",legend=False)
plt.title("Movies vs TV Shows")
plt.xlabel("Type")
plt.ylabel("Count")
plt.show()
```

# Key Insights:
1. Movies are available in a much higher number than TV Shows on Netflix.
2. This indicates that Netflix's content library is dominated by movies.

```python
plt.figure(figsize=(14,6))
df_counts=df['release_year'].value_counts().sort_index()
# df_counts.columns=['release_year','count']
# df_counts=df.counts.sort_values('release_year')
sns.lineplot(x=df_counts.index,y=df_counts.values,marker="o",markerfacecolor="red",markeredgecolor="red",color="green")

plt.title("Content Released by Year")
plt.xlabel("Release Year")
plt.ylabel("Number of Titles")
plt.show()
```

#  Key Insights:
1. The number of Netflix titles generally increased over the years, showing continuous growth in content production.
2. Recent years have a higher number of releases compared to older years.


```python
plt.figure(figsize=(10,5))
top_countries=df["country"].value_counts().head(10)
sns.barplot(x=top_countries.index,y=top_countries.values,color="#59A14F")
plt.xticks(rotation=45)
plt.title("Top 10 Countries With Most Netfliex Content")
plt.xlabel("Country")
plt.ylabel("Number of Titles")
plt.show()
```

#  Key Insights:
1. The United States has the highest number of Netflix titles.
2. Other countries contribute fewer titles compared to the United States.


```python
# Split multiple genres and count each genre
genres = df['listed_in'].str.split(', ', expand=True).stack().value_counts().head(10)

plt.figure(figsize=(10,5))
sns.barplot(x=genres.values, y=genres.index,palette="viridis",hue=genres,legend=False)

plt.title("Top 10 Genres on Netflix")
plt.xlabel("Number of Titles")
plt.ylabel("Genre")
plt.show()
```

#  Key Insights:
1. International Movies have the highest number of titles on Netflix, followed by Dramas and Comedies, showing that these genres are the most popular on the platform.
2. Netflix offers a wide variety of genres, but drama and international content dominate its library, catering to a global audience.

```python
plt.figure(figsize=(10,5))

sns.countplot(
    y='rating',
    data=df,
    order=df['rating'].value_counts().index,palette="Set1",hue="rating",legend=False
)

plt.title("Distribution of Netflix Content Ratings")
plt.xlabel("Number of Titles")
plt.ylabel("Rating")
plt.show()
```

#  Key Insights:
1. TV-MA is the most common content rating on Netflix, indicating that a large portion of the content is intended for mature audiences.
2. Other ratings such as TV-14 and TV-PG also have a significant number of titles, showing that Netflix provides content for different age groups.

```python
director_df = df[df['director'] != 'Not Available']
top_directors=director_df["director"].value_counts().head(10)
plt.figure(figsize=(10,5))
sns.barplot(x=top_directors.values,y=top_directors.index,palette="Paired",hue=top_directors,legend=False)
plt.title("Top 10 Directors on Netflix")
plt.xlabel("Number of Titles")
plt.ylabel("Director")
plt.show()
```

#  Key Insights:
1. The top directors have contributed multiple movies and TV shows to Netflix, showing their strong presence on the platform.
2. The chart highlights the directors with the highest number of titles available on Netflix.

```python
df['date_added'] = pd.to_datetime(df['date_added'], format='mixed')
df['year_added'] = df['date_added'].dt.year
```

```python
df[['date_added', 'year_added']].head()
```

```python
plt.figure(figsize=(10,5))
counts_by_year=df['year_added'].value_counts().sort_index()
sns.lineplot(x=counts_by_year.index,y=counts_by_year.values,marker="o",color="green")
plt.xticks(rotation=45)
plt.title("Content Added to Netflix by Year")
plt.xlabel("Year Added")
plt.ylabel("Number of Titles")
plt.show()
```

#  Key Insights:
1. The number of titles added to Netflix increased steadily from 2015 and reached its peak in 2019, showing rapid growth in Netflix's content library.
2. After 2019, the number of titles added decreased slightly in 2020 and 2021, but Netflix continued to add a large amount of content.


```python

movies = df[df['type'] == 'Movie'].copy()

movies['duration_min'] = movies['duration'].str.extract(r'(\d+)')

movies = movies.dropna(subset=['duration_min'])

movies['duration_min'] = movies['duration_min'].astype(int)
```

```python
movies['duration_min'].head()
```

```python
plt.figure(figsize=(12,6))
sns.histplot(data=movies,x='duration_min', bins=30,color="teal",kde=True)
plt.title("Movie Duration Distribution")
plt.xticks(rotation=0)
plt.xlabel("Duration (Minutes)")
plt.ylabel("Number of Movies")
plt.show()
```

#  Key Insights:
1. Most Netflix movies have a duration between 80 and 120 minutes,with the highest number around 90-100 minutes. 
2. Very short and very long movies are less common compared to medium-length movies.

```python

top_duration = movies['duration'].value_counts().head(10)

plt.figure(figsize=(10,5))

sns.barplot(x=top_duration.index, y=top_duration.values,color="Pink")

plt.title("Top 10 Movie Durations on Netflix")
plt.xlabel("Duration")
plt.ylabel("Number of Movies")
plt.xticks(rotation=45)
plt.show()
```

#  Key Insights:
1. Most Netflix movies have durations between 90 and 100 minutes, making them the most common movie lengths.
2. Movies with very short or very long durations are less common compared to standard-length movies.
